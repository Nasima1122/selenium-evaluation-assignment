```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
```

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://www.naukri.com/")
```

```python
designation=driver.find_element(By.CLASS_NAME,"suggestor-input ")
designation.send_keys('Data Analyst')
```

```python
search=driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()
```

```python
job_details=[]
job_location=[]
company_name=[]
experience_required=[]
```

```python
details=driver.find_elements(By.XPATH,'//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div[1]')
for i in details:
    job_details.append(i.text)
    
location=driver.find_elements(By.XPATH,'//span[@class="ni-job-tuple-icon ni-job-tuple-icon-srp-location loc"]')
for i in location:
    job_location.append(i.text)
    
name=driver.find_elements(By.XPATH,'//div[@class=" row2"]/span/a[1]')
for i in name:
    company_name.append(i.text)
    
required=driver.find_elements(By.XPATH,'//span[@class="ni-job-tuple-icon ni-job-tuple-icon-srp-experience exp"]')
for i in required:
    experience_required.append(i.text)
```

```python
print(len(job_details),len(job_location),len(company_name),len(experience_required))
```

```python
df=pd.DataFrame({'Details':job_details,'Location':job_location,'Name':company_name,'Required':experience_required})
df
```
