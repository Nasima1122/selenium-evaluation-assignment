```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
```

```python
driver= webdriver.Chrome()
```

```python
driver.get("https://www.flipkart.com/apple-iphone-15-black-128-gb/p/itm6ac6485515ae4?")
```

```python
rating_phone=[]
review_title=[]
review_detail=[]
```

```python
start=0
end=10
for page in range(start,end):

    phone=driver.find_elements(By.XPATH,'//div[@class="_3LWZlK _1BLPMq"]')
    for i in phone:
        rating_phone.append(i.text)
    title=driver.find_elements(By.XPATH,'//p[@class="_2-N8zT"]')
    for i in title:
        review_title.append(i.text)
    detail=driver.find_elements(By.XPATH,'//div[@class="t-ZTKy"]')
    for i in detail:
        review_detail.append(i.text)
    
   
    
    
```

```python
print(len(rating_phone),len(review_title),len(review_detail))
```

```python
df=pd.DataFrame({'Phone':rating_phone,'Title':review_title,'Detail':review_detail})
df
```
